```python
def cols_90per_nulls(data):
    count = 0
    cols_to_drop = {}
    for col in data.columns: 
        per_nulls = data[col].isna().sum()/len(data[col])
        if per_nulls >= 0.5:
            cols_to_drop[col] = per_nulls 
            # print(col, per_nulls)
            count+=1
        else:
            None
    
    print('Number of cols with >50% nulls:', count)
    return cols_to_drop
```

# Revision de columnas con valores nulos (>= 80%)

```python
nulls_file = {}
for file in dict_files:
    filename = directorio+dict_files[file]
    df_temp = pd.read_csv(filename)
    nulls_file[file] = cols_90per_nulls(df_temp)
```

```python
for idx, pais in enumerate(nulls_file):
    if idx == 0:
        df_tmp = pd.DataFrame(data = nulls_file[pais].values(), index= nulls_file[pais].keys(), columns=[pais])
    else:
        df = pd.DataFrame(data = nulls_file[pais].values(), index= nulls_file[pais].keys(), columns=[pais])
        df_tmp = pd.concat([df_tmp, df], axis=1)

df_tmp
```

```python
df_cols = pd.DataFrame( df_tmp.T.ge(0.8).all() )
cols_to_drop = df_cols[df_cols[0] == True].index.to_list()
len(cols_to_drop)
```

los archivos comparten las mismas columnas con mas del 80% de los datos nulos....se almacena el nombre de las columnas en la lista cols_to_drop

## Separando Columnas

```python
#modificar el formato de la variable 'Estimated Revenue Range'
df['Estimated Revenue Range'] = df['Estimated Revenue Range'].apply(lambda elem: str(elem).replace('$', ''))
```

```python
dfSplit = df['Headquarters Location'].str.split(",", n = 3, expand = True) 
```

```python
# creacion de columna Income
df["Ciudad"]= dfSplit[0] 
# creacion de columna Type
df["Departamento"]= dfSplit[1] 
# creacion de columna Description
df["Pais"]= dfSplit[2]
```

```python
#Funcion que corrige espacios 
def correct_word(word):
    
    new_word = word.split()[0]
    return new_word
```

```python
#Aplicandon la funcion para la columna Departamento
df['Departamento'] = df['Departamento'].apply(correct_word)
```

```python
#Aplicandon la funcion para la columna Pais
df['Pais'] = df['Pais'].apply(correct_word)
```

```python
#Cambiar el formato de las variable
df['Last Funding Amount']=pd.to_numeric(df['Last Funding Amount'])
df['Last Equity Funding Amount']=pd.to_numeric(df['Last Equity Funding Amount'])
```
