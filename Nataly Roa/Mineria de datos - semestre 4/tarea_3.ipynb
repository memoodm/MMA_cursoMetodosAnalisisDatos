#  Eliminar colunas con 75% de nulidad en las variables

```python
def cols_75per_nulls(data):
    count = 0
    cols_to_drop = {}
    for col in data.columns: 
        per_nulls = data[col].isna().sum()/len(data[col])
        if per_nulls >= 0.75:
            cols_to_drop[col] = per_nulls 
            # print(col, per_nulls)
            count+=1
        else:
            None
    
    print('Number of cols with >75% nulls:', count)
    return cols_to_drop
```
