## PREGUNTA 1: ¿De los diversos factores de predicción elegidos inicialmente afectan realmente a la esperanza de vida? ¿Cuáles son las variables de predicción que realmente afectan la esperanza de vida?

```python
corrMatrix = df.corr()
LIFE_EXPECTANCY_POSS = 1

# se filtran los valores que no se encuentren superando el rango de 0.4
colum_names = []
columns_values = []
for column in corrMatrix.columns:
    if(abs(corrMatrix.iloc[LIFE_EXPECTANCY_POSS][column])>0.4):
        colum_names.append(column)
        columns_values.append(corrMatrix.iloc[LIFE_EXPECTANCY_POSS][column])

plt.barh(colum_names,columns_values)
plt.grid()
```

<h3 style="color: red"> Respuesta: <h3>
<p>Todas las variables afectan la "expectativa de vida" ya que ninguna de sus correlaciones es de 0, pero algunas por supuesto afectan esta mas que otras, por tanto se tomaran principalmente las que superen el rango de 0.5<p>
    
<p>Negativamente: </p> thinness 5-9 years, thinness 1-19 years, HIV/AIDS, Adult Mortality

    
<p>Positivamente:</p> Schooling, Income  composition of resources, GDP, Diphtheria, Polio, BMI, Alchool 


# PREGUNTA 2: ¿Debería un país con un valor de esperanza de vida inferior (<65) aumentar su gasto sanitario para mejorar su esperanza de vida media?


```python
lessThan65 = df.loc[(df['Life expectancy '] < 65)]['Total expenditure'].mean()
greatherThan65 = df.loc[(df['Life expectancy '] >= 65)]['Total expenditure'].mean()

print("Gasto sanitario promedio de los paises que tienen un promedio de vida inferior a 65: %.6f"%(lessThan65))
print("Gasto sanitario promedio de los paises que tienen un promedio de vida superior a 65: %.6f"%(greatherThan65))
```

<h3 style="color: red"> Respuesta: </h3>
SI, segun los datos ya que esto SI mejora en promedio la esperanza de vida de su poblacion

# PREGUNTA 4: ¿Tiene la esperanza de vida una correlación positiva o negativa con los hábitos alimenticios, el estilo de vida, el ejercicio, el tabaquismo, el consumo de alcohol, etc.?
