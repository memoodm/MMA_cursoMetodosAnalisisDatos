```python
import math
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate

pi= np.pi
X = np.arange(0, 1, 0.01)
N = 10

def Uxt(x,t,coeficientes):
    suma = 0
    e = math.e
    for k in range(1,N):
        try:
            A = coeficientes[k]
            B = math.cos(k*pi*t)
            C = math.sin(k*pi*x)
            suma += A*B*C
        except:
            continue
    return 4/(pi**2)*suma

def u0x(x):
    if x >= 0 and x <= 1/4:
        return 4*x
    if x >= 1/4 and x <= 1:
        return (4/3)*(1-x)

def u0x_(x,k):
    return u0x(x)*math.sin(k*pi*x)
    
coeficientes = []
for n in range(1,N):
    value1, error1 = scipy.integrate.quad(lambda x: u0x_(x,n), 0, 1/4)
    value2, error2 = scipy.integrate.quad(lambda x: u0x_(x,n), 1/4, 1)
    coeficientes.append(value1+value2)


times = np.arange(0,1,0.1)
for t in times:
    U = []
    u = []
    for x in X:
        U.append(Uxt(x,t,coeficientes))
        u.append(u0x(x))
    
    plt.suptitle("For t = "+str(t))
    plt.plot(X, U,label = 'Serie de Fourier')
    plt.plot(X, u,label = 'u(0,x)')
    plt.legend()
    plt.grid()
    plt.show()
```
